```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
sns.set()
```

```python
acc_train = pd.read_csv('train_cleaned.csv',index_col=False)
acc_train.drop(columns=['Unnamed: 0'],inplace=True)
acc_train.head(30)
```

```python
drop_again =['Start_Time',"End_Time", "Turning_Loop",'Street', 'County', 'Zipcode','Start_Time','Description','Wind_Direction',
            'Railway', 'Give_Way','No_Exit','Traffic_Calming','Bump','Turning_Loop','Roundabout']
X = acc_train.drop(columns=drop_again, inplace=False)
```

```python
##################################################################
#ignore
```

```python
list_values = ['City','State','Weather_Condition','day_of_Week','month','year','Hour']
num_vals = [25, 25,15,6,12,5,24]
```

```python
for i in range(len(list_values)):
    col_name = list_values[i]
    num_cols = num_vals[i]
    col_df = pd.get_dummies(X[col_name],prefix=col_name, prefix_sep='_')
    col_df = col_df.iloc[:,0:num_cols]
    X = X.join(col_df)
```

```python
X.drop(columns = list_values, inplace=True)
```

```python
bool_vals = ['Amenity','Crossing',
               'Junction','Station','Stop','Traffic_Signal',
             'Sunrise_Sunset']

for i in bool_vals:
    X[i] = X[i] * 1
```

```python
col_df = pd.get_dummies(X['Side'],prefix="Side", prefix_sep='_')
X = X.join(col_df)
X.drop(columns=['Side'],inplace=True)
```

```python
X.head()
```

```python
X.to_csv("Forest_data.csv")
```
